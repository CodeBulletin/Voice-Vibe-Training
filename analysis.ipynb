```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Load the data

```python
df = pd.read_csv('datasets_v4/dataset_1d.csv')
df.head()
```

```python
# plot the emotions distribution

emotions = df['Emotion'].value_counts()
emotions.plot.bar()
plt.show()
```

```python
# plot intensity distribution

intensity = df['Intensity'].value_counts()
intensity.plot.bar()
plt.show()
```

```python
# Merge emotions fear and fearfull, calm and neutral

df['Emotion'] = df['Emotion'].replace('fearful', 'fear')
df['Emotion'] = df['Emotion'].replace('calm', 'neutral')
```

```python
# plot the emotions distribution

emotions = df['Emotion'].value_counts()
emotions.plot.bar()
plt.show()
```
